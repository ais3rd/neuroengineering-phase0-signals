Objective:
Generate a discrete-time sine wave, add Gaussian noise,
and visualize the result.

```python
import numpy as np
import matplotlib.pyplot as plt

duration = 2.0
num_samples = 1000
sampling_rate = num_samples / duration

f = 0.5  # Hz
```

```python
t = np.linspace(0, duration, num_samples)
y_clean = np.sin(2 * np.pi * f * t)

plt.figure(figsize=(8,4))
plt.plot(t, y_clean)
plt.title("Clean Sine Wave (0.5 Hz)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
np.random.seed(0)

noise_amplitude = 0.3
noise = noise_amplitude * np.random.randn(len(t))
y_noisy = y_clean + noise

plt.figure(figsize=(8,4))
plt.plot(t, y_clean, label="Clean")
plt.plot(t, y_noisy, label="Noisy", alpha=0.7)
plt.legend()
plt.grid(True)
plt.show()
```
